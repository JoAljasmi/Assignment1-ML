### Adding imports

```python
import math
import random
import numpy as np
```

# Making a neuron

```python
class Neuron:
    def __init__(self, n_inputs: int):
        # random weights and bias
        self.weights = [random.uniform(-1, 1) for _ in range(n_inputs)]
        self.bias = random.uniform(-1, 1)

    # putting in activation functions
    def sigmoid(self, x: float) -> float:
        return 1 / (1 + math.exp(-x))
    
    def relu(self, x: float) -> float:
        return max(0.0, x)

    def leaky_relu(self, x: float, alpha: float = 0.01) -> float:
        return x if x > 0 else alpha * x

    def tanh(self, x: float) -> float:
        return math.tanh(x)


# forward passes

    def forward(self, inputs: list, activation: str = "sigmoid") -> float:

        z = sum(x * w for x, w in zip(inputs, self.weights)) + self.bias

        activations = {
            "sigmoid": self.sigmoid,
            "relu": self.relu,
            "leaky_relu": self.leaky_relu,
            "tanh": self.tanh
        }
        return activations[activation](z)

    
    #tester
if __name__ == "__main__":
    neuron = Neuron(n_inputs=3)
    output = neuron.forward([0.5, -0.3, 0.2], activation="sigmoid")
    print(f"Output: {output:.4f}")
    print(f"Weights: {[f'{w:.3f}' for w in neuron.weights]}")
    print(f"Bias: {neuron.bias:.3f}")
    
```

# making a neuron with numpy

```python
class Neuron:
    def __init__(self, n_inputs: int):
        # random weights and bias
        self.weights = np.random.uniform(-1, 1, size=n_inputs)
        self.bias = random.uniform(-1, 1)

    # putting in activation functions
    def sigmoid(self, x: float) -> float:
        return 1 / (1 + math.exp(-x))
    
    def relu(self, x: float) -> float:
        return max(0.0, x)

    def leaky_relu(self, x: float, alpha: float = 0.01) -> float:
        return x if x > 0 else alpha * x

    def tanh(self, x: float) -> float:
        return math.tanh(x)


# forward passes

    def forward(self, inputs, activation: str = "sigmoid") -> float:

        z = np.dot(self.weights, inputs) + self.bias

        activations = {
            "sigmoid": self.sigmoid,
            "relu": self.relu,
            "leaky_relu": self.leaky_relu,
            "tanh": self.tanh
        }
        return activations[activation](z)

    
    #tester
if __name__ == "__main__":
    neuron = Neuron(n_inputs=3)
    output = neuron.forward([0.5, -0.3, 0.8], activation="relu")
    print(f"Output: {output:.4f}")
    
```

```python
class DenseLayer:
    def __init__(self, n_inputs: int, n_neurons: int):
        self.weights = np.random.uniform(-1, 1, size=(n_neurons, n_inputs))
        self.biases = np.random.uniform(-1, 1, size=n_neurons)

    def sigmoid(self, x):
            return 1 / (1 + np.exp(-x))

    def relu(self, x):
            return np.maximum(0, x)

    def forward(self, inputs, activation: str = "sigmoid"):
            inputs = np.array(inputs)

            z = self.weights @ inputs + self.biases

            if activation == "sigmoid":
                return self.sigmoid(z)
            elif activation == "relu":
                return self.relu(z)
            else:
                return z

if __name__ == "__main__":
    layer = DenseLayer(n_inputs=784, n_neurons=128)
    random_image = np.random.uniform(0, 1, size=784)
    output = layer.forward(random_image, activation="relu")
    print(f"Output shape: {output.shape}")
    print(f"first 5 outputs: {output[:5].round(4)}")
```
